<h1> Object tracking using multiple trackers </h1>

```python
import cv2
import os
import sys
```

```python

print("Select your choice of tracker: ","\nTrackers available: ","\n1.Boosting","\n2.MIL","\n3.KCF","\n4.GOTURN")
trackers = ["BOOSTING","MIL","KCF","GOTURN"]
```

```python

num = int(input("Select tracker number\n"))

if(num==1):
    tracker = cv2.TrackerBoosting_create()
    tracker_name = "BOOSTING"
    print("Selected Tracker: ",tracker_name)
elif(num==2):
    tracker = cv2.TrackerMIL_create()
    tracker_name= "MIL"
    print("Selected Tracker: ",tracker_name)
elif(num==3):
    tracker = cv2.TrackerKCF_create()
    tracker_name = "KCF"
    print("Selected Tracker: ",tracker_name)
elif(num==4):
    tracker = cv2.TrackerGOTURN_create()
    tracker_name = "GOTURN"
    print("Selected Tracker: ",tracker_name)
else:
    tracker = cv2.TrackerGOTURN_create() 
    tracker_name = "GOTURN"
    print("Switching to default tracker: ",tracker_name)
```

```python
video = cv2.VideoCapture("test_video.mp4")
video.set(cv2.CAP_PROP_FPS, 20)
fps = video.get(cv2.CAP_PROP_FPS)  # to get the FPS of the video
print("FPS is :", fps)

# Exit if video not opened
if not video.isOpened():
    print("Could not open video")
    sys.exit()

#reading frames
ok,frame = video.read()
if not ok:
    print("Cannot read video file")
    sys.exit()
    
#bbox = (276, 23, 86, 320)
bbox = cv2.selectROI(frame,False)
ok = tracker.init(frame,bbox)
delayTime = 50
while True:
    # Read a new frame

    ok, frame = video.read()
    if not ok:
        break
 
    # Start timer
    timer = cv2.getTickCount()
 
    # Update tracker
    ok, bbox = tracker.update(frame)
    print("Bbox are: ",bbox)
    
 
    # Calculate Frames per second (FPS)
    fps = cv2.getTickFrequency() / (cv2.getTickCount() - timer);
 
    # Draw bounding box
    if ok:
        # Tracking success
        p1 = (int(bbox[0]), int(bbox[1]))
        p2 = (int(bbox[0] + bbox[2]), int(bbox[1] + bbox[3]))
        cv2.rectangle(frame, p1, p2, (255,0,0), 2, 1)
    else :
        # Tracking failure
        cv2.putText(frame, "Object out of frame", (100,80), cv2.FONT_HERSHEY_SIMPLEX, 0.75,(0,0,255),2)
 
    # Display tracker type on frame
    cv2.putText(frame, tracker_name, (100,20), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (50,170,50),2);
 
    # Display FPS on frame
    cv2.putText(frame, "FPS : " + str(int(fps)), (100,50), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (50,170,50), 2);
 
    # Display result
    cv2.imshow("Tracking", frame)
  
    # Exit if ESC pressed
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break
cv2.destroyAllWindows()
```
